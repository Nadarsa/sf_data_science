```python
#9.10. Переведите числовой признак CreditScore в категориальный.
#Для этого воспользуйтесь функцией get_credit_score_cat(), которая приведена ниже.
#Примените её к столбцу CreditScore и создайте новый признак CreditScoreCat — категории кредитного рейтинга.
#Постройте сводную таблицу, строками которой являются категории кредитного рейтинга (CreditScoreCat),
#а столбцами — количество лет, в течение которых клиент пользуется услугами банка (Tenure).
#В ячейках сводной таблицы должно находиться среднее по признаку оттока (Exited) — доля ушедших пользователей.

#На основе полученной сводной таблицы постройте тепловую карту с аннотацией.
#Найдите на тепловой карте категории клиентов, которые уходят чаще всего.

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 851:
        return "Top"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score < 300:
        return "Deep"

churn_data = pd.read_csv('data/churn.csv')
churn_data = churn_data.drop(labels='RowNumber', axis=1)
churn_data['CreditScoreCat'] = churn_data['CreditScore'].apply(get_credit_score_cat)

#display(churn_data.head())
#print(churn_data.info())

pivot = churn_data.pivot_table(
    values='Exited',
    columns='Tenure',
    index='CreditScoreCat',
)
pivot.columns = pivot.columns.astype('string')
#display(pivot)

heatmap = sns.heatmap(data=pivot, cmap='YlGnBu', annot=True)
heatmap.set_title('Тепловая карта доли оттока клиентов', fontsize=16);

print('Анализируя тепловую карту, можно заметить, что чаще всего уходит категория клиентов с кредитным рейтингом Very_Poor. \
Так, в данной категории визуально выделяется наибольшее количество ярких ячеек, соответствующих долям ушедших пользователей. \
Также чуть менее яркие ячейки можно увидеть в категориях Excellent и Poor. \
Большая доля ушедших клиентов это категория Very_Roor с продолжительностью пользования услугами банка до года (0,44). \
Далее идет всё так же категория, но количество лет составляет, по порядку убывания доли, 10 (0,33), 3 (0,29), 4 (0,29), 1 (0,25), 8 (0,25) лет. \
Что касается категории Excellent, то здесь наибольшие доли ушедших клиентов составляют, соответственно по порядку убывания, \
до года (0,28), 9 (0,27) и 10 лет (0,25). В категории Poor доли ушедших клиентов пользовались банковскими услугами в течение 1 года (0,26) и 9 лет (0,25). \
Итак, чаще всего уходят следующие категории клиентов: Very_Poor, Excellent и Poor, - так как именно в этих категориях \
встречаются наиболее яркие ячейки, то есть наибольшие величины долей ушедших клиентов.')
```
