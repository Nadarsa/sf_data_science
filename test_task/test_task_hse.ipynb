```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data.xlsx')
data.head(5)
```

```python
data.info()
```

```python
data['Дата начала программы'] = pd.to_datetime(data['Дата начала программы'])
data['Дата окончания программы'] = pd.to_datetime(data['Дата окончания программы'])
```

```python
data[data['Страна'] != 'Российская Федерация'].groupby('Страна')['Дата начала программы'].count().sort_values(ascending=False)
```

```python
data[data['Страна'] != 'Российская Федерация'].groupby('Страна')['Дата начала программы'].count().sort_values(ascending=False).head(10)
```

```python
data['Длительность программы в днях'] = pd.to_timedelta(data['Дата окончания программы'] - data['Дата начала программы'])
```

```python
data.head(3)
```

```python
data.groupby('Наименование подразделения')['Длительность программы в днях'].mean().sort_values(ascending=False)
```

```python
data.groupby('Наименование подразделения')['Длительность программы в днях'].mean().sort_values(ascending=False).head(5)
```

```python
mask1 = data['Наименование принимающей организации'] != 'НИУ ВШЭ'
mask2 = data['Тип организации'] == 'образовательная организация высшего образования'
mask3 = data['Тип организации'] == 'научная организация'

academic_mob_df = data[mask1 & (mask2 | mask3)]
academic_mob_df.head(3)
academic_mob_df.groupby('Наименование подразделения')['Дата начала программы'].count().sort_values(ascending=False).head(5)
```
